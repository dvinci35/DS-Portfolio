#### A Function to automate - averaging all the Level % category-wise to know the performance of students in a particular year based on the category

```python
oFile.Category.astype('category')
oFile.Year.astype('category')

# making a function for automating the average
def fu(yr):
    data_yr = oFile[oFile['Year']==yr]
    data_yr_avg =data_yr.groupby(['Category'])[[
        'Level 1 %',
        'Level 2 %',
        'Level 3 %',
        'Level 4 %']].sum()/7
    print(f"This is data_yr_avg {data_yr_avg}")
    # plotting the graph
    gr = data_yr_avg.plot(kind='bar', figsize=(10,8),
                title=f'Performance of students in {yr}')
    return gr
```

#### Defining a Function that will gather data by grade-wise and average the percent scores, plot that graph

```python
def grade_comp(grade):
    data_grade= oFile[oFile['Grade']==grade]
    # print(data_grade.head(6))
    data_std_year = data_grade.groupby(['Year'])[['Level 1 %','Level 2 %','Level 3 %','Level 4 %', ]].sum()/4
    # print(data_std_year.head(6))
    
    return data_std_year.plot(kind='bar', figsize=(10,8), title=f"Performance of Grade {grade} students.")
```

### Performance of Grade 4 students from 2006 to 2011

```python
grade_3 = grade_comp(3)
plt.ylabel("Percentage of Total Students")
```

#### Conclusion: There is overall increase in students scoring Level 1. Students scoring Level2 drastically increased in 2010 and 2011. The students scoring Level3 was unusually high in 2009 and drastically dropped and was least in 2009. The students scoring Level4 were least in 2011.
